## Исходные данные:
- **данные сервиса Яндекс Недвижимость — архив объявлений за несколько лет о продаже квартир в Санкт-Петербурге и соседних населённых пунктах**

- **О каждой квартире в базе содержится два типа данных: добавленные пользователем и картографические. Например, к первому типу относятся площадь квартиры, её этаж и количество балконов, ко второму — расстояния до центра города, аэропорта и ближайшего парка**c

**СТРУКТУРА ТАБЛИЦЫ С ИСХОДНЫМИ ДАННЫМИ(СТОЛБЦЫ):**

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)

### Тип этажа квартиры (значения — «первый», «последний», «другой»)

```python
def floors_type(row):
    floor = row['floor']
    total_floors = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == total_floors:
        return 'последний'
    else:
        return 'другой'
```

#### Проверка гипотезы

**Построили график корреляции**

```python
data.plot(x='total_area', y='last_price', kind='scatter', alpha=0.3)
```

**Т.к в нижнем левом углу точки сильно сливаются, поэтому построили** *hexagonal binning plot*

```python
data.plot(x='total_area', y='last_price', kind='hexbin', gridsize=20, \
        figsize=(8, 6), sharex=False, grid=True, ylim=(-20, 50), xlim=(0, 150))
plt.show()
```

**Данный график малоинформативен. Поэтому нашли коэффициент корреляции**

```python
data['last_price'].corr(data['total_area'])
```

**Из коэффициента корреляции видно, что зависимость действительно существует. Но, т.к в разных районах стоимость одного квадратного метра может очень сильно отличается, следует вычислить коэффициент корреляции для каждого района в отдельности и вычислить их медиану**

```python
data_names = data['locality_name'].unique()
#подготовили сводную таблицу для коэффициентов
total_count_data = data.pivot_table(index='locality_name', \
                                   values='total_area', aggfunc='count')
total_area_data = pd.DataFrame(data= data_names, \
                                columns=['locality_name'])   
total_area_data = total_area_data.merge(total_count_data,\
                                        on='locality_name', how='left')
total_area_data.rename(columns = {'total_area' : 'count'}, inplace=True)
total_area_data = total_area_data.query('count > 20')

total_area_data
```

**Написали и применили функцию для вычисления коэффициентов корреляции. Для оценки оставили только те населенные пункты, где количетво квартир более 20**

```python
def total_area_corr(row):
    data_row = data.query('locality_name == @row')
    return round(data_row['last_price'].corr(data_row['total_area']), 2)
```

```python
total_area_data['corrs'] = total_area_data['locality_name'].apply(total_area_corr)
total_area_data
```

**Из получившейся таблицы видно, что для большинства районов коэффициенты корреляции составляют более 0,8.
Для большей объективности исследования посчитали их медианное значение**

```python
total_area_data['corrs'].median()
```

#### Вывод

**Гипотеза полностью подтвердилась и по всем населенным пунктам вместе, и по каждому в отдельности(медианное значение коэффициента корреляции по всем населенным пунктам оказалось больше 0,85).Тот факт, что коэффициент корреляции по всему исходному датасету оказался сильно меньше, чем медиана коэффициентов для каждого района можно объяснить разной стоимостью одного квадратного метра в разных населенных пунктах, что сильно искажает зависимость по всему исходному датасету**
